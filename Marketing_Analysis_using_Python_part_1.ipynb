```python
#check data type of column

marketing['sessions'].dtype

#change the data type of a column


marketing['sessions'] = marketing['sessions'].astype('bool')


#To check data type value is changed from int64 to boolean, print marketing['ga:sessions']

marketing['sessions'].head()
```

```python
#Creating a new boolean column if channel value is 'Direct'

marketing['is_channel_Direct'] = np.where(marketing['channel'] == 'Direct', True, False)

#To check new column boolean values on the basis of 'Direct' Channel column
print(marketing['is_channel_Direct'].head(3))
```

```python
#To make a dictionary of marketing_channels

channel_dict = {"Direct" : 1, "Organic" : 2, "Affiliates" : 3, "Facebook" : 4}

marketing['channel_code'] = marketing['channel'].map(channel_dict)

print(marketing['channel_code'].head(3))
```

```python
#Read date column using parse_date

marketing = pd.read_excel("C:/Users/New User/Desktop/nnnt.xlsx", parse_dates = ['date'])

#or Convert already existing column to datetime column

marketing['date'] = pd.to_datetime(marketing['date'])

#Check day of week from date column

marketing['dayofweek'] = marketing['date'].dt.dayofweek

#Here 0 maps to Monday, 1 maps to Tuesday
marketing['dayofweek'].head(3)
```

```python
#Aggregate unique users that see ads by date


daily_users = marketing.groupby(['date'])['sessions'].nunique()


daily_users
```

```python
#Visualizing Results

import matplotlib.pyplot as plt

#Plot

daily_users.plot()
```

```python
#Annotate

plt.title('Daily Numbers of users who see ads')
plt.xlabel('Date')
plt.ylabel('Number of Users')
plt.xticks(rotation = 45)
plt.show()
```
