```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import time
```

```python
#Read csv
fruit_sales = pd.read_excel('fruit.xlsx')

#print(fruit_sales.head(2))


#plot bargraph using seaborn barplot

sns.barplot(x = 'name', y = 'quantity', data = fruit_sales) #To plot bar graph
plt.show() #To display bar graph
plt.savefig('awesome_plot.png') #To save bargraph in png format


#Don't use plt.show() and plt.savefig() simultaneously, It will save only blank png file
```

```python
#To plot graph with better version

sns.set_style('whitegrid') #To add style in Bar Graph

sns.barplot(x = 'name', y = 'quantity', data = fruit_sales) #To plot bar graph

plt.title('Quantity of Fruits') #Title of Bar Graph

plt.xlabel('Name') #Title of X - axis

plt.ylabel('Quantity') #Title of Y - axis

sns.despine() #To remove unwanted borders

plt.show() #To display bar graph
```

```python
#read excel

movie_data = pd.read_excel('movie theater sales data.xlsx')

movie_data.head(3)


#To plot graph with hue (for color legend)

sns.set_style('whitegrid') #To add style in Bar Graph

sns.barplot(x = 'movie_title', y = 'ticket_quantity', data = movie_data, hue = 'ticket_type') #To plot bar graph
plt.legend(bbox_to_anchor=(1, 1))

plt.title('No. of Movie Tickets of Movie Title') #Title of Bar Graph

plt.xlabel('Movie Title') #Title of X - axis

plt.ylabel('No. of Tickets') #Title of Y - axis
plt.xticks(rotation=60) #Rotation of X - Axis Title
plt.show() #To display Bar graph
```
