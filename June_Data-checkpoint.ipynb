```python
import pandas as pd
```

```python
june_data = pd.read_excel("D:/Office Work/NB_Walkin_Analysis/June_Final_Overlap_v1.xlsx")
june_data = june_data.reset_index()
june_data = june_data.drop_duplicates(subset=['Phone_No_Model_ID','Channel']).reset_index(drop=True)
june_data.head()
```

```python
june_data_pivot = pd.crosstab(june_data.Phone_No_Model_ID,june_data.Channel)
june_data_pivot = june_data_pivot.reset_index()
june_data_pivot_slice = june_data_pivot.loc[(june_data_pivot['GOOGLE_SEARCH_NON_BRAND'] >= 1) & (june_data_pivot['WALKIN'] >= 1)]
```

```python
june_data_pivot_slice.to_excel("D:/Office Work/NB_Walkin_Analysis/June_Final_Final_Overlap_v1.xlsx", index=False)
```
