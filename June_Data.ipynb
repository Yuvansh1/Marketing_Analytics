```python
import pandas as pd
```

```python
april_data = pd.read_excel("D:/Office Work/NB_Walkin_Analysis/03-07-2019/June_Final_Overlap_new_v1.xlsx")
april_data = april_data.reset_index()
april_data = april_data.drop_duplicates(subset=['Phone_No','Channel']).reset_index(drop=True)
april_data.head()
```

```python
april_data_pivot = pd.crosstab(april_data.Phone_No,april_data.Channel)
april_data_pivot = april_data_pivot.reset_index()
april_data_pivot_slice = april_data_pivot.loc[(april_data_pivot['GOOGLE_SEARCH_NON_BRAND'] >= 1) & (april_data_pivot['WALKIN'] >= 1)]
```
